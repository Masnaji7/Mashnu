```python
for c in cols :
    q1 = dfcar[c].quantile(0.25)
    q3 = dfcar[c].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - (1.5 * iqr)
    upper = q3 + (1.5 * iqr)
    dfcar[c] = dfcar[c].map(lambda x : lower if x < lower else  upper if x > upper else x)
```

```python
scorelist = []
for i in range(500):
    xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.20,random_state = i)
    lreg = LinearRegression()
    lreg.fit(xtrain,ytrain)
    sc = lreg.score(xtest,ytest)
    scorelist.append(sc)

m = max(scorelist)
print(m)
print(scorelist.index(m))

```
