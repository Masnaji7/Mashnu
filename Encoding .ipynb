### One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
ohe
```

```python
ohe.fit(dfr[["gender"]])
```

```python
ohe.transform(dfr[["gender"]])
```

```python
ohe.transform(dfr[["gender"]]).toarray()
```

```python
ary1 = ohe.transform(dfr[["gender"]]).toarray()
```

```python
ary1
```

```python
pd.DataFrame(ary1,dtype = "int")
```

```python
ohe.categories_
```

```python
pd.DataFrame(ary1 , dtype = "int" , columns = ['Female', 'Male'])
```

```python
dfen = pd.DataFrame(ary1 , dtype = "int" , columns = ['Female', 'Male'])
```

```python
dfen
```

```python
df = pd.concat([dfr,dfen],axis = 1)
```

```python
df
```

### Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ord = OrdinalEncoder()
```

```python
ord
```

```python
ord.fit(dfr[["education"]])
```

```python
ord.transform(dfr[["education"]])
```

```python
ord.categories_
```

```python
ary2 = ord.transform(dfr[["education"]])
```

```python
ary2
```

```python
df["education"] = ary2.astype("int")
```

```python
df
```
